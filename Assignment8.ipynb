```python
def array_properties(a):
    print("a = \n", a)
    print("dim = ", a.ndim)
    print("shape = ", a.shape)
    print("datatype = ", a.dtype)
    print("size = ", a.size)
    print()
```

```python
#1. Create a 3D data of shape 10x5x3 witn values equal zero

import numpy as np

zeros_arr = np.zeros((10, 5, 3))

array_properties(zeros_arr)
```

```python
#2. Create another array with value 255 and shape 10x5x3

full_arr = np.full((10, 5, 3), 255)
array_properties(full_arr)
```

```python
#3. Join the 2 array horizontally

arr = np.concatenate((zeros_arr, full_arr))
print(arr)
```

```python
#4. Create an array of random unsigned integer 8 bit with shape(50x50x3)

np.random.seed(11)

a1 = np.random.randint(0, 256, size=(50, 50, 3),dtype= 'uint8')
array_properties(a1)
```

```python
import matplotlib.pyplot as plt

plt.imshow(a1)
plt.title('a1')
plt.show()
```

```python
a1[:26,:,:] = [255,0,0]
a1[26:50,:,:] = [0,255,0]

plt.imshow(a1)
plt.title('red & green')
plt.show()
```
