```python
# 1. Create an array of shape (100,100,3) of unisigned integer 8 bits
#    fill with random values between 0-256.
# 2. Find the values that are less than 30,50,100,250.
# 3. Split the array into 4 parts vertically and horizontally.
# 4. Copy array from question 1 and change the values that are 
#    less than 118 to 0 and others to 255.
# 5. Hint use random seed of 12345
```

```python
def array_properties(a):
    print("a = \n", a)
    print("dim = ", a.ndim)
    print("shape = ", a.shape)
    print("datatype = ", a.dtype)
    print("size = ", a.size)
    print()
```

```python
#question 1 & 5

import numpy as np

np.random.seed(12345)

a1 = np.random.randint(0,256,(100,100,3),dtype='uint8')
array_properties(a1)
```

```python
#2.
iboolles30 = a1 <= 30 
iboolles50 = a1 <= 50 
iboolles100 = a1 <= 100
iboolles250 = a1 <= 250

iboolles30 = iboolles30 == True
iboolles50 = iboolles50 == True
iboolles100 = iboolles100 == True
iboolles250 = iboolles250 == True

les30result = a1[iboolles30]
array_properties(les30result)

les50result = a1[iboolles50]
array_properties(les50result)

les100result = a1[iboolles100]
array_properties(les100result)

les250result = a1[iboolles50]
array_properties(les250result)
```

```python
#4.
newa1=a1.copy()
array_properties(newa1)
print(newa1)
```

```python
ibool180 = newa1 <=180
array_properties(ibool180)

iboolles180 = ibool180 == True 
iboolmore180 = ibool180 == False 

newa1[iboolles180]=0
newa1[iboolmore180]=255

array_properties(newa1)
```

```python
#3.
a1_hsp = np.hsplit(a1,4)
print(a1_hsp)
```
