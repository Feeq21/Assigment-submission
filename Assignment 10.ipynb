```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

try:
    img = cv.imread("..\\essential\\assets\\klcc.JPG",cv.IMREAD_COLOR)
    
    h, w, c = img.shape
    print('width:  ', w)
    print('height: ', h)
    print('channel:', c)
    print('size (height * width): ', img.size)
# display file
    #cv.imshow(f'CA: {img.shape}', img)

# pause execution here by waiting for a user to press a key
    #k = cv.waitKey(0)

#add border 10%
    tw = w * 0.1
    borderSizeTw = int(tw)
    
    th = h * 0.1
    borderSizeTh = int(th)
    
#copy image
    img_corners = img.copy()
    
#top   
    img_corners[:borderSizeTw,:borderSizeTh] = [255,255,255]
    img_corners[:borderSizeTw,-borderSizeTh:] = [255,255,255]
    
#bottom
    img_corners[-borderSizeTw:,-borderSizeTh:] = [255,255,255]
    img_corners[-borderSizeTw:,:borderSizeTh] = [255,255,255]
     
    cv.imshow('4 square',img_corners)
    k = cv.waitKey(0)
    
except Exception as ex:
    print(ex)
    
finally:
    cv.imwrite("White four corners.png", img_corners)
    cv.destroyAllWindows()
```
